This notebook isolates methodology sections from all papers in Zootaxa, EJT and the Pensoft journals in our corpus. We skipped Insectamundi because papers in that journal often did not seem to have separate methods sections. After isolating these methods, we split them into paragraphs and into sentences. We will use these sections, paragraphs and sentences for separate topic-models in order to 1) explore the methods in the corpus, 2) design a classification scheme for these methods, and 3) identify the right level of granularity (full paper, methods section, paragraph or sentence) to design a classifier for methods in taxonomy papers.  

This notebook uses functions from the isolate_methods_functions.py file

```python
# get the methods from each non-insectamundi paper, and write them to a separate text file

error = {}
no_methods = []
methods_papers = []

# loop over all files in the taxonomy corpus except for the IM ones
all_filenames_noIM = [i for i in all_filenames if 'InsectaMundi' not in i]

for i in tqdm(all_filenames_noIM):


    # check if there is a methodsfile
    try:

        # get the methods section
        methods = get_methods(i)

        # store papers without methods
        if methods == None:
            no_methods.append(i)
            
        else:

            # store papers with methods in a list of filenames
            methods_papers.append(i)

            # write the methods file
            filepath = r"C:\Users\conix\Documents" + i[1:].split('.json')[0] + "_methods.txt"
            with open(filepath, "w", encoding = 'utf8') as text_file:
                print(methods, file=text_file)

    # store papers that throw an error and check these
    # these are 6 pensoft papers that have differently structured xmls
    except Exception as e:
        error[i] = e
        continue

    
        
```

## Random samples

```python
rs_marlies2 = 123
rs_laura2 = 456
rs_stijn3 = 789

random.seed(rs_stijn3)

# Function to get n random key-value pairs from a dictionary
def get_random_kv_pairs(dct, n, excluded_keys):
    if n > len(dct):
        raise ValueError("n is larger than the number of elements in the dictionary.")
        
    available_keys = [key for key in dct.keys() if key not in excluded_keys]
    
    # Get n random keys
    random_keys = random.sample(available_keys , n)

    
    
    # Create a new dictionary with the random key-value pairs
    random_kv_pairs = {key: dct[key] for key in random_keys}
    
    return random_kv_pairs

# Specify the number of random pairs you want
n = 300  # Change this number as needed

# say which keys should be excluded
keys = pd.read_csv(r'C:\Users\conix\Dropbox\FNRS project taxonomy\methods in taxonomy\coding_trial\firstTry\corrected_dfx.csv', sep = ';')
excluded_keys = keys.id.values

# Get n random key-value pairs
random_methods_paras = get_random_kv_pairs(methods_paras, n, excluded_keys)

output_file_path = r'C:\Users\conix\immigration_framing\data_files\stijn_samples3.json'


# Write the random key-value pairs to the output file
with open(output_file_path, 'w') as f:
    for key, text in random_methods_paras.items():
        entry = {
            "id": key,
            "text": text,
            "annotations": []
        }
        f.write(json.dumps(entry) + '\n')

print(f"Output written to {output_file_path}")
```

```python
def read_jsonfile(filepath):
    data = []
    try:
        with open(filepath, 'r', encoding='utf-8') as file:
            for line in file:
                try:
                    data.append(json.loads(line))
                except json.JSONDecodeError:
                    print(f"Error decoding line: {line}")
    except FileNotFoundError:
        print(f"The file {filepath} was not found.")
    return data

filepath1 = "..."


data = read_jsonfile(filepath1)
```

## Brute search

```python
# dct with the labels and key words we will search for
# add more if we notice our model has trouble with some categories

search_dct = {

    "chemo"      :  ['chemotaxonomy', 'chemical', 'metabolites', 'pheromone', 'metabolomics', 'phytochemistry'],
    "ML"         :  ['supervised','unsupervised','neural net'],
    "Proteomic"  :  ['allozyme','proteomic',"karyoteype",'cytological'],
    "Molecular_analysis"  :  ['bayesian','gmyc','ptp','coalescent','neighbour','parsimony','supertree'],
    "Molecular_data"  :  ['electrophoresis','microsatellite','epigenetic','illumina','nanopore', 'RAPD', 'SNP'],
    "Spec_just"       :  ['criterion','reproductive isolation','reproductively isolated','species concept','Species Concept','BSC','PSC','arbitrary'],
    "sequencing"      :  ['nanopore','solexa','pacbio','ion torrent',"illumina"],
    "genotype_nonphylo" : ["structurama","baps","dapc"],
    "phylo_SD"        : ["gmyc","ptp", "spedestem","bp&p", 'snapp', "mptp", "bayes factor"],
    "phylo_tree"      : ["supertree","consensus tree", 'consensustree']
    

}

result_dct = {key: [] for key in search_dct}


# Iterate over each document and search term list
for doc_id, text in methods_paras.items():
    for category, words_list in search_dct.items():
        # Check if any word in the words_list is present in the document text
        if any(word in text.lower() for word in words_list):
            result_dct[category].append(doc_id)
        
```

```python
# select n from each label

search_selected = {}

n = 15

for i in result_dct.values():
    if len(i) > 14:
        sample = random.sample(i, n)
        for doc in sample:
            search_selected[doc] = methods_paras[doc]
    else:
        sample = random.sample(i, len(i))
        for doc in sample:
            search_selected[doc] = methods_paras[doc]

```
